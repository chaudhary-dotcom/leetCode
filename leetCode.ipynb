### 1. Two Sum 

##### Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

```python
nums = [2,7,11,15]

target = 9

for i in range(len(nums)):
    for j in range(i+1, len(nums)):
        if nums[i] + nums[j] == target:
            print([i, j])
```

```python
def twosum(nums, target):
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i,j]

print(twosum([2,7,11,15], 9))
print(twosum([3,2,4],6))
print(twosum([3,8,4,6],7))
```

### 2. Palindrome Number

```python
# convert int to string
def isPalindrome(n):
    return str(n) == ''.join(reversed(str(n)))

n = 10

if isPalindrome(n):
    print('The number is palindrome: ')
else:
    print('The number is not palindrome')
```

```python
# without convert int into the string
def isPalindrome(a):
    if a < 0:
        return False
    
    original = a
    reverse = 0

    while a > 0:
        last_digit = a % 10 
        reverse = reverse*10+last_digit
        a = a//10
    
    if original == reverse:
        return True
    
    else:
        return False
    
isPalindrome(33)
```

```python
# list of palindrome
palindromes = []
for i in range(1,50,1):
    reverse = 0
    original = i
    
    while i > 0:
        last_digit = i%10
        reverse = reverse*10+last_digit
        i = i//10
    if original == reverse:
        palindromes.append(original)

print(palindromes)
```

### 1. Integer to Roman

```python
class solution:
    def intToRoman(self, num: int) -> str:   # this -> str means: takes integer and retruns a string
        
        val_map = [
            (1000, 'M'), (900, 'CM'),
            (500, 'D'), (400, 'CD'),
            (100, 'C'), (90, 'XC'),
            (50, 'L'), (40, 'XL'),
            (10, 'X'), (9, 'IX'),
            (5, 'V'), (4, 'IV'),
            (1, 'I')
        ]

        roman = ''

        for value, symbol in val_map:
            while num >= value:
                roman += symbol
                num -= value
        
        return roman
    
slt = solution()
slt.intToRoman(1900)
```

#### 1. Longet Common Prefiix

```python
pre = ['flower', 'flow','f']

pre.sort()

first = pre[0]
last = pre[-1]

prefix = ''

for i in range(min(len(first), len(last))):
    if first[i] == last[i]:
        prefix += first[i]
    else:
        break
    print(last[i])
print(prefix)

```

```python
class Solution:
    def longestCommonPrefix(self, strs):

        if not strs:
            return ''
        strs.sort()

        first = strs[0]
        last = strs[-1]

        prefix = ''
        for i in range(min(len(first), len(last))):
            if first[i] == last[i]:
                prefix += first[i]
            
            else:
                return prefix
        return prefix

so = Solution()
print(so.longestCommonPrefix(['flower', 'flow', 'flight']))
```
